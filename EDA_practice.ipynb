```python
import matplotlib.pyplot as plt
import seaborn as sns

# Apply seaborn theme
sns.set_style("darkgrid")

x = [11, 34, 76, 46, 91]
y = [10, 20, 30, 40, 50]

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', linestyle='--', marker='o', linewidth=2, markersize=8)
plt.title("Stylish Line Plot", fontsize=16, color='navy')
plt.xlabel("Data 1", fontsize=12)
plt.ylabel("Data 2", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

data = pd.DataFrame({
    "class": ['a','b','c','d','e'],
    "student": [67, 46, 35, 67, 55]
})#palette="pastel",

plt.figure(figsize=(7, 5))
sns.barplot(x='class', y='student', data=data,  edgecolor='black')
plt.title("Stylish Bar Plot", fontsize=16)
plt.xlabel("Class", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
data2 = pd.DataFrame({
    'class': ['a', 'b', 'a', 'c', 'a', 'b', 'c', 'c', 'c', 'd', 'd', 'e']
})

plt.figure(figsize=(7, 5))
sns.countplot(x='class', data=data2, palette="Set2", edgecolor='black')
plt.title("Stylish Count Plot", fontsize=16)
plt.xlabel("Class", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

data3 = np.random.randn(100)

plt.figure(figsize=(8, 5))
plt.hist(data3, bins=10, color='skyblue', edgecolor='black')
plt.title("Stylish Histogram", fontsize=16)
plt.xlabel("Value Range", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data3, shade=True, color="purple", linewidth=2)
plt.title("Stylish KDE Plot", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(y=data3, color='lightgreen', linewidth=2, width=0.3)
plt.title("Stylish Box Plot", fontsize=16)
plt.grid(True, axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# load dataset
df=sns.load_dataset('iris')
df
#pair plot
sns.pairplot(df,hue='species')
plt.show()
```

```python
#heatmap
df_numeric=df.drop(columns='species')
corelation_matrix=df_numeric.corr()
corelation_matrix
sns.heatmap(corelation_matrix,annot=True,cmap='coolwarm')
plt.title("corelation")
plt.show()
```

```python
#pie chart
labels=['A','B','C']
sizes=[30,40,30]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=0)
plt.title("pie chart")
plt.show()
```

```python
sns.swarmplot(x='species',y='sepal_length',data=df)
plt.title("swarm chart")
plt.show()
```

```python
#strip plot
sns.stripplot(x='species',y='sepal_length',data=df,jitter=True)
plt.title("strip plot")
plt.show()
```

```python
#joint plot
sns.jointplot(x='sepal_length',y='petal_length',data=df,kind='scatter')
plt.show()
```
