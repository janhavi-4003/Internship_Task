```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds
```

```python
ds.info()
```

# Categorical variable Analysis

```python
#Employee count per Department
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title('Employee count per Department')
```

Finance department having more employee count followed by marketing and engineering 

```python
# Gender Distribution
sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title('Gender Distribution')
```

Female employees having more in count compared to male employee

```python
#Employee Location Distribution
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title('Number of Employees by Location')
plt.xticks(rotation=45)
```

seattle location having more count of employee followed by New York and San francisco

```python
# Age Distrubution
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title(' Age Distrubution')
```

Between 45 to 50 aged(expreienced) employees are more count followed by 25-30 aged (Freshers) Employee

```python
# years of experience Distrbution
sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title('years of experience')
```

2.5 to 5.0 years of experience employees are more

```python
# monthly salary Distribution
sns.boxplot(data=ds,y='Monthly_Salary',color='lightblue')
plt.title('salary Spread')
```

Each deparment give minimum 20,000 and maximum 1,20,000 salary and averagr salary is 75,000.

```python
#Average salary by Department
sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette='viridis')
plt.title('Average salary by Department')
```

Finance deparment employee has more salary

```python
#performance score by department
sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title('performance score by department')
```

marketing department having maximum performance score than other department


```python
# salary distribution by gender
sns.violinplot(data=ds,x='Gender',y='Monthly_Salary',palette='muted')
plt.title('salary distribution by gender')
```

Male employees having more salary as compered to female employee

```python
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
#Salary  vs Experience Scatter plot
sns.scatterplot(data=ds,x='Years_of_Experience',y='Monthly_Salary',hue='Department')
sns.title('Salary  vs Experience Scatter plot')
```

 HR Department having maximum salary and more year of experience than other department
